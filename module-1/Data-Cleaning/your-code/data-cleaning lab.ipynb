```python
#convert data types
data = data.astype('object')
#find all null values
null_cols = data.isnull().sum()
print(null_cols)
#create new dataframe by dropping columns with over 70% null values
# data.shape[0]*.7
drop_cols = list(null_cols[null_cols > data.shape[0]*.7].index)
ndf = data.drop(drop_cols, axis = 1)
# confirm columns dropped
# ndf.shape
```

```python


ndf.columns = map(colFormat, ndf)

ndf_uniq = [[i, len(ndf[i].unique())] for i in ndf.columns]
ndf_uniq.sort(key = lambda x: x[1])
#print(ndf_uniq)
#print((ndf['CASE NUMBER.1'] == ndf['CASE NUMBER.2'])& (ndf['CASE NUMBER.1'] == ndf['CASE NUMBER']).value_counts())
#print((ndf['HREF FORMULA'] == ndf['HREF']).value_counts())
dro_col = ['CASE NUMBER.1','CASE NUMBER.2','HREF FORMULA']
ndf1 = ndf.drop(dro_col, axis = 1)
ndf1['SEX'].value_counts()
ndf1['SEX'] = list(map(colFormat, ndf1['SEX']))


ndf1['SEX'] = list(map(unifColx, ndf1['SEX']))
ndf1['FATAL (Y/N)'] = list(map(unifCold, ndf1['FATAL (Y/N)']))

for i in range(len(ndf1['NAME'])):
    if 'FEMALE' in str(ndf1['NAME'][i]).upper():
        ndf1['SEX'][i] = 'F'
        ndf1['NAME'][i] = 'NaN'
    if 'MALE' in str(ndf1['NAME'][i]).upper():
        ndf1['SEX'][i] = 'M'
        ndf1['NAME'][i] = 'NaN'
    if 'GIRL' in str(ndf1['NAME'][i]).upper():
        ndf1['SEX'][i] = 'F'
        ndf1['NAME'][i] = 'NaN'
    if 'BOY' in str(ndf1['NAME'][i]).upper():
        ndf1['SEX'][i] = 'M'
        ndf1['NAME'][i] = 'NaN'
    if 'anonymous' in str(ndf1['NAME'][i]).lower():
        ndf1['NAME'][i] = 'NaN'
        
ndf1['NAME'] = list(map(nTS, ndf1['NAME']))
drop3 = ['CASE NUMBER','NAME', 'PDF', 'HREF','ORIGINAL ORDER','INVESTIGATOR OR SOURCE','LOCATION','AREA']
ndf2 = ndf1.drop(drop3, axis = 1).drop_duplicates()
ndf2.head()

act_lst = pd.DataFrame(ndf2['ACTIVITY'].value_counts())
cvt_act_lst = list(act_lst[act_lst['ACTIVITY'] == 1].index)
ndf2['ACTIVITY'] = list(map(clnAct, ndf2['ACTIVITY']))
ndf2['AGE'] = list(map(clnAge, ndf2['AGE']))
ndf2
```
