## Challenge 2: Count Commits
Count how many commits were made in the month of october of 2018.
#### 1. Obtain all the commits made in October 2018 via API, which is a JSON array that contains multiple commit objects.

`````python
# your code here
username = 'Skye-FY'
token = '66c7c3ad8f94a76cb4e1116988e8d96e2e981e78'
gh_session = requests.Session()
gh_session.auth = (username, token)
url = 'https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/commits'
commits = gh_session.get(url).json()
print(commits)

# commits[2]['commit']['committer']['date']
`````

#### 2. Count how many commit objects are contained in the array.

`````python
# your code here
def count_commits(data, time_range_start, time_range_end):
    time_range = DateTimeRange(time_range_start, time_range_end)
    commit_time_range = [data[i]['commit'] for i in range(len(data)) if data[i]['commit']['committer']['date'] in time_range]
    return len(commit_time_range)

oct18 = count_commits(commits, '2018-10-01T00:00:00Z', '2018-10-31T23:59:59Z')
since_oct18 = count_commits(commits, '2018-10-01T00:00:00Z', '2020-02-13T23:59:59Z')
print('%d commits made during October 2018; %d commits made since October 2018.' % (oct18, since_oct18))
`````

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. #### 1. Find the secret files.

`````python
# your code here
username = 'Skye-FY'
token = '66c7c3ad8f94a76cb4e1116988e8d96e2e981e78'
gh_session = requests.Session()
gh_session.auth = (username, token)
url = 'https://api.github.com/repos/ironhack-datalabs/scavenger'
scavenger = gh_session.get(url).json()
scavenger
`````

`````python
url_s = 'https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1'
scavenger_tree = gh_session.get(url_s).json()
print(scavenger_tree)
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here
def find_info(data, info):
    return list([data[i][info] for i in range(len(data)) if data[i][info] != None])

# scc = find_info(scavenger_tree['tree'],'url')

# scavenger_tree
# scavenger_tree['tree']

scc_list = find_info(scavenger_tree['tree'],'path')
scl = [''.join(['https://raw.githubusercontent.com/ironhack-datalabs/scavenger/master/',i]) for i in scc_list if 'scavengerhunt' in i]
sorted_scl = sorted(scl, key = lambda x: x.split('.')[3])
sorted_scl
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:

`````python
# your code here
username = 'Skye-FY'
token = '66c7c3ad8f94a76cb4e1116988e8d96e2e981e78'
gh_session = requests.Session()
gh_session.auth = (username, token)
sc_txt = list(map(lambda x: gh_session.get(x).text.strip('\n)'), sorted_scl))
print(sc_txt)
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here
sc_con = ' '.join(sc_txt)
`````
