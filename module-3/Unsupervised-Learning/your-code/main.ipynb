```python
'''
Descriptive statistics - any outliers to remove?

'''
data_num = data[data.columns[2:]]
fig, ax = plt.subplots(1,6,figsize=(120,20))
for i in range(len(data_num.columns)):
    sns.boxplot(data_num[data_num.columns[i]],ax=ax[i])
```

```python
data_num.describe()
q1 = []
q3 = []
iqr = []
rule_l = []
rule_u = []
outliers = []
for i in data_num.describe().columns:
    q1.append(data_num.describe()[i]['25%'])
    q3.append(data_num.describe()[i]['75%'])
    iqr.append(float(data_num.describe()[i]['75%']-data_num.describe()[i]['25%']))
for i in range(len(iqr)):
    rule_l.append(float(q1[i]-iqr[i]*1.5))
    rule_u.append(float(q3[i]+iqr[i]*1.5))
    outliers.append(list(data_num[data_num[data_num.describe().columns[i]]<rule_l[i]][data_num.columns[i]])+list(data_num[data_num[data_num.describe().columns[i]]>rule_u[i]][data_num.columns[i]]))

```

```python
data_num = data[data.columns[2:]].reset_index()
fig, ax = plt.subplots(1,6,figsize=(120,20))
for i in range(len(data_num.columns)-1):
    sns.scatterplot(x='index', y=data_num.columns[1+i],data=data_num,ax=ax[i])
```

```python
'''
Column-wise data distribution - is the distribution skewed?

-- all skewed
'''
data_num = data[data.columns[2:]]
fig, ax = plt.subplots(1,6,figsize=(120,20))
for i in range(len(data_num.columns)):
    sns.distplot(data_num[data_num.columns[i]],ax=ax[i])
```
